```python
import numpy as np
import pandas as pd
import time
import random as rand
import matplotlib.pyplot as plt

from SalesConverging import TSPSolver as Standard
from SalesOptimized import TSPSolver as Optimized
```

```python
#city route data frame after solving, file holding initial city data, file_name to store graph in
def plot_routes(cities_after, cities_file, file_name = None):
    cities_before = pd.read_csv(cities_file, 
                 sep='\s+',
                 comment='#',            # Ignore lines starting with #
                 names=['Longitude', 'Latitude', 'City'],  # Column names
                 quotechar='"')
    cities_before = cities_before[["Longitude", "Latitude"]]
    
    longitude_range = [min(cities_before['Longitude'])-5, max(cities_before['Longitude'])+5]
    latitude_range = [min(cities_before['Latitude'])-5, max(cities_before['Latitude'])+5]


    # Create plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    # Plot cities and route
    ax1.plot(cities_before['Longitude'], cities_before['Latitude'],
            marker='x', color='blue', linewidth=2,
            markersize=8, linestyle='-')

    ax2.plot(cities_after['Longitude'], cities_after['Latitude'],
            marker='x', color='red', linewidth=2,
            markersize=8, linestyle='-')

    # Set plot parameters
    ax1.set_title("Salesman's Cities Before ")
    ax1.set_xlabel("Longitude")
    ax1.set_ylabel("Latitude")
    ax1.set_xlim(longitude_range)
    ax1.set_ylim(latitude_range)

    ax2.set_title("Salesman's Cities After")
    ax2.set_xlabel("Longitude")
    ax2.set_ylabel("Latitude")
    ax2.set_xlim(longitude_range)
    ax2.set_ylim(latitude_range)
    
    if(file_name):
        plt.savefig(file_name)
        plt.close()
    else:
        plt.show()
```

# Comparison of two methods on data set of 23 cities

```python
control = Standard('cities.dat')
start_distance = control.total_distance()

start_time = time.time()
cooling_rate = .99 
n_trials =20
max_bad = 100

solver1 = Standard('cities.dat', cooling_rate, n_trials, max_bad)
df1, temps1, distances1 = solver1.solve()
end_time = time.time()
elapsed_time1 = end_time - start_time

distance1 = solver1.total_distance()

start_time = time.time()

max_iterations = 1500   
initial_temp = 15.0  
cooling_rate = .001 
ntrials = 50
solver2 = Optimized('cities.dat', max_iterations, initial_temp, cooling_rate, ntrials)
df2, temps2, distances2 = solver2.solve()
end_time = time.time()

elapsed_time2 = end_time - start_time
distance2 = solver2.total_distance()

print(f'Start distance with random route: {start_distance: .2f} km')
print(f'Standard: Distance: {distance1: .2f} km, Time: {elapsed_time1: .2f} s')
print(f'Optimized: Distance: {distance2: .2f} km, Time: {elapsed_time2: .2f} s')
```

### Plots for Standard Solution

```python
df1_sorted = df1.sort_index()
df1_sorted = df1_sorted.drop(columns='Initial')
plot_routes(df1_sorted, 'cities.dat')

plt.plot(temps1, distances1)
plt.title("Temperature vs Total Distance")
plt.gca().invert_xaxis()
plt.show()
```

### Plots for Optimized Solution

```python
df2_sorted = df2.sort_index()
df2_sorted = df2_sorted.drop(columns='Initial')
plot_routes(df2_sorted, 'cities.dat')

plt.plot(temps2, distances2)
plt.title("Temperature vs Total Distance")
plt.gca().invert_xaxis()
plt.show()
```

# Optimized Method on data set of 1207 cities
The standard method takes too long to converge due to the size of the problem so only the optimized method is shown in this notebook. Even with the decreased time of the optimized method, the time to achieve the best improvement is too large so parameters are selected to insure the simulation runs in a reasonable time and makes some progress.

```python
start_time = time.time()

max_iterations = 100000   
initial_temp = 25.0   
cooling_rate = .0001
ntrials = 200

solver = Optimized('cities3.dat', max_iterations, initial_temp, cooling_rate, ntrials)
df, temps, distances = solver.solve()
end_time = time.time()

elapsed_time = end_time - start_time
distance = solver.total_distance()

print(f'Start distance with random route: {start_distance: .2f} km')
print(f'Standard: Distance: N/A, Time: N/A')
print(f'Optimized: Distance: {distance: .2f} km, Time: {elapsed_time: .2f} s')
```

```python
df_sorted = df.sort_index()
df_sorted = df_sorted.drop(columns='Initial')
plot_routes(df_sorted, 'cities3.dat')

plt.plot(temps, distances)
plt.title("Temperature vs Total Distance")
plt.gca().invert_xaxis()
plt.show()
```
